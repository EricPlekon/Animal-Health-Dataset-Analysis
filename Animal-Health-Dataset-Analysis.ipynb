```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# reading csv file
df = pd.read_csv('Health_AnimalBites.csv')
```

```python
# outputs the first 5 rows of the dataframe
df.head()
```

## Lets plot this data species in a bar graph

```python
# bar graph
pd.value_counts(df['SpeciesIDDesc']).plot(kind='barh')
```

```python
# check for missing values
df.isnull().sum()
```

```python
# summary statistics
df.describe()
```

## Exploratory Data Analysis

```python
# distribution of animal types invovled in bites

plt.figure(figsize=(10, 6))
sns.countplot(y='SpeciesIDDesc', data=df, order=df['SpeciesIDDesc'].value_counts().index)
plt.title('Distribution of Animal Types Involved in Bites')
plt.xlabel('Count')
plt.ylabel('Animal Species')
plt.show()
```

```python
# Gender Distribution of Animals involved in bites

plt.figure(figsize=(10, 6))
sns.countplot(y='GenderIDDesc', data=df, order=df['GenderIDDesc'].value_counts().index)
plt.title('Gender Distribution of Animals Involved in Bites')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.show()
```

```python
# Location of Bites on Body

plt.figure(figsize=(12, 8))
sns.countplot(y='WhereBittenIDDesc', data=df, order=df['WhereBittenIDDesc'].value_counts().index)
plt.title('Location of Bites on the Body')
plt.xlabel('Count')
plt.ylabel('Body Part')
plt.show()
```

```python
# Age of Vaccination 
plt.figure(figsize=(10, 6))
sns.histplot(df['vaccination_yrs'].dropna(), bins=30, kde=True)
plt.title('Distribution of Vaccination Age')
plt.xlabel('Years Since Vaccination')
plt.ylabel('Frequency')
plt.show()
```
